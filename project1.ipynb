```python
#년도별 한중일 소비자 물가지수
import csv
f = open('sss.csv')
data = csv.reader(f)
max_date = []
kor =[]
cha =[]
jpn =[]

name1 = '대한민국'
name2 = '중국'
name3 = '일본'
name4 = '국가별'

for row in data:
    if name1 in row[0]:
        for i in row[1:27:]:
            kor.append(float(i))
    elif name2 in row[0]:
        for i in row[1:27:]:
            cha.append(float(i))
    elif name3 in row[0]:
        for i in row[1:27:]:
            jpn.append(float(i))
    
    elif name4 in row[0]:
        for i in row[1:27:]:
            max_date.append(int(i))       
                       
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 3), dpi = 300)
plt.plot(max_date, kor, label = 'kor')
plt.plot(max_date, jpn, label = 'jap')
plt.plot(max_date, cha, label = 'cha')
plt.title('년도별 한중일 소비자 물가지수')
plt.legend()
plt.savefig('년도별 한중일 소비자 물가지수.png', facecolor='#eeeeee') #저장
plt.show()
```

```python
#년도별 한중일 소비자 물가지수 상승률
import csv
f = open('sss.csv')
data = csv.reader(f)
max_date = []

kor =[]
cha =[]
jpn =[]

kor_price_growth=[]
cha_price_growth=[]
jpn_price_growth=[]

name1 = '대한민국'
name2 = '중국'
name3 = '일본'
name4 = '국가별'

for row in data:
    if name1 in row[0]:
        for i in row[1:27:]:
            kor.append(float(i))   
    elif name2 in row[0]:
        for i in row[1:27:]:
            cha.append(float(i))
    elif name3 in row[0]:
        for i in row[1:27:]:
            jpn.append(float(i))
            
    elif name4 in row[0]:
        for i in row[2:27:]:
            max_date.append(int(i))       

for i in range(0,25):
    kor_price_growth.append(kor[i+1]/kor[i])
    cha_price_growth.append(cha[i+1]/cha[i])
    jpn_price_growth.append(jpn[i+1]/jpn[i])

import matplotlib.pyplot as plt
plt.figure(figsize = (10, 3), dpi = 300)
plt.plot(max_date, kor_price_growth, label = 'kor_price_growth')
plt.plot(max_date, cha_price_growth, label = 'cha_price_growth')
plt.plot(max_date, jpn_price_growth, label = 'jpn_price_growth')
plt.title('년도별 한중일 소비자 물가지수 상승률')

plt.legend()
plt.savefig('년도별 한중일 소비자 물가지수 상승률.png', facecolor='#eeeeee') #저장
plt.show()
```
